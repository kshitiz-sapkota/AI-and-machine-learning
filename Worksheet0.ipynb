Task - 1:
Create a Python program that converts between different units of measurement.
- The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.
- Requirements:
1. Functions: Define at least one function to perform the conversion.
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).
3. User Input: Prompt the user to select the conversion type and input the value.
4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and return value.
- Conversion Options:
1. Length: Convert meters (m) to feet (ft) / feet (ft) to meters (m).
2. Weight: Convert kilograms (kg) to pounds (lbs) / pounds (lbs) to kilograms (kg).
3. Volume: Convert liters (L) to gallons (gal) / gallons (gal) to liters (L).

```python
def convert_units(conversion_type, value):
    conversions = {
        "m_to_ft": value * 3.28084,
        "ft_to_m": value / 3.28084,
        "kg_to_lbs": value * 2.20462,
        "lbs_to_kg": value / 2.20462,
        "l_to_gal": value * 0.264172,
        "gal_to_l": value / 0.264172
    }

    if conversion_type in conversions:
        return conversions[conversion_type]
    else:
        raise ValueError("Unsupported conversion type.")

def unit_conversion_program():
    print("Select conversion type:")
    print("1. m_to_ft")
    print("2. ft_to_m")
    print("3. kg_to_lbs")
    print("4. lbs_to_kg")
    print("5. l_to_gal")
    print("6. gal_to_l")

    choice_map = {
        "1": "m_to_ft",
        "2": "ft_to_m",
        "3": "kg_to_lbs",
        "4": "lbs_to_kg",
        "5": "l_to_gal",
        "6": "gal_to_l"
    }

    try:
        choice = input("Enter choice (1-6): ")
        if choice not in choice_map:
            raise ValueError("Invalid choice.")

        value = float(input("Enter value to convert: "))
        result = convert_units(choice_map[choice], value)

        print(f"Converted value: {result}")

    except ValueError as e:
        print("Error:", e)

unit_conversion_program()
```

**Task - 2:**

Create a Python program that performs various mathematical operations on a list of numbers.
- The Program should:
1. Prompt the user to choose an operation (e.g., find the sum, average, maximum, or minimum of the numbers).
2. Ask the user to input a list of numbers (separated by spaces).
3. Perform the selected operation and display the result.
4. Handle potential errors, such as invalid input or empty lists.
- Requirements:
1. Functions: Define at least one function for each operation (sum, average, maximum, minimum).
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values or empty lists).
3. User Input: Prompt the user to select the operation and input the list of numbers.
4. Docstrings: Include a docstring in each function to describe its purpose, parameters, and return value.

```python
def calculate_sum(numbers):
    return sum(numbers)

def calculate_average(numbers):
    if len(numbers) == 0:
        raise ValueError("List is empty.")
    return sum(numbers) / len(numbers)

def find_maximum(numbers):
    if len(numbers) == 0:
        raise ValueError("List is empty.")
    return max(numbers)

def find_minimum(numbers):
    if len(numbers) == 0:
        raise ValueError("List is empty.")
    return min(numbers)

def math_operations_program():
    print("Choose operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = input("Enter choice (1-4): ")
        numbers = list(map(float, input("Enter numbers separated by spaces: ").split()))

        if not numbers:
            raise ValueError("List cannot be empty.")

        if choice == "1":
            print("Result:", calculate_sum(numbers))
        elif choice == "2":
            print("Result:", calculate_average(numbers))
        elif choice == "3":
            print("Result:", find_maximum(numbers))
        elif choice == "4":
            print("Result:", find_minimum(numbers))
        else:
            print("Invalid choice.")

    except ValueError:
        print("Error: Please enter valid numeric input.")

math_operations_program()
```

```python
def extract_every_other(lst):
    return lst[::2]

def get_sublist(lst, start, end):
    return lst[start:end+1]

def reverse_list(lst):
    return lst[::-1]

def remove_first_last(lst):
    return lst[1:-1]

def get_first_n(lst, n):
    return lst[:n]

def get_last_n(lst, n):
    return lst[-n:]

def reverse_skip(lst):
    return lst[-2::-2]

data = [10, 21, 32, 43, 54, 65, 76, 87, 98, 109]

every_other = extract_every_other(data)
first_five = get_first_n(data, 5)
last_three = get_last_n(data, 3)
middle_part = remove_first_last(data)
sub = get_sublist(middle_part, 1, 4)
reversed_sub = reverse_list(sub)
weird_backward = reverse_skip(data)

print("Original:         ", data)
print("Every other:      ", every_other)
print("First 5:          ", first_five)
print("Last 3:           ", last_three)
print("Without first & last:", middle_part)
print("Sublist(1:4) from middle:", sub)
print("Reversed sublist: ", reversed_sub)
print("reverse_skip:     ", weird_backward)
```

**4.3 Exercise on Nested List:**

1. Flatten a Nested List: flatten(lst)
2. Accessing Nested List Elements: access_nested_element(lst, indices)
3. Sum of All Elements in a Nested List: sum_nested(lst)
4. Remove Specific Element from a Nested List: remove_element(lst, elem)
5. Find the Maximum Element in a Nested List: find_max(lst)
6. Count Occurrences of an Element in a Nested List: count_occurrences(lst, elem)
7. Flatten a List of Lists of Lists: deep_flatten(lst)
8. Nested List Average: average_nested(lst)

```python
def flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(item)
        else:
            result.append(item)
    return result

def access_nested_element(lst, indices):
    element = lst
    for index in indices:
        element = element[index]
    return element

def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result

def find_max(lst):
    maximum = float('-inf')
    for item in lst:
        if isinstance(item, list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum, item)
    return maximum

def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

def average_nested(lst):
    flat_list = deep_flatten(lst)
    if len(flat_list) == 0:
        raise ValueError("List is empty.")
    return sum(flat_list) / len(flat_list)

nested = [1, 2, [3, 4, [5, 5, 6]], [7, [8, 9]], 10, [11, [12]], 5]

flat_one = flatten(nested)
print("flatten (one level):", flat_one)

val = access_nested_element(nested, [2, 2, 1])
print("access_nested_element([2,2,1]):", val)

total = sum_nested(nested)
print("sum_nested:", total)

cleaned = remove_element(nested, 5)
print("remove_element 5:", cleaned)

mx = find_max(nested)
print("find_max:", mx)

five_count = count_occurrences(nested, 5)
print("count_occurrences of 5:", five_count)

completely_flat = deep_flatten(nested)
print("deep_flatten:", completely_flat)

avg = average_nested(nested)
print("average_nested:", avg)
```

**Problem - 4: Matrix Operations:**

For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);

Prove the following with Numpy:
1. Prove A.A⁻¹ = I.
2. Prove AB ≠ BA.
3. Prove (AB)ᵀ = BᵀAᵀ.

Solve the following system of Linear equations using Inverse Methods:

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}

Now: solve the above equation using np.linalg.inv function.

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1. Prove A.A^-1 = I
A_inv = np.linalg.inv(A)
print("A * A_inv:\n", np.dot(A, A_inv))

# 2. Prove AB ≠ BA
print("AB:\n", np.dot(A,B))
print("BA:\n", np.dot(B,A))

# 3. Prove (AB)^T = B^T A^T
left = np.dot(A,B).T
right = np.dot(B.T, A.T)
print("(AB)^T:\n", left)
print("B^T A^T:\n", right)
```

```python
A = np.array([[2,-3,1],
              [1,-1,2],
              [3,1,-1]])

B = np.array([-1,-3,9])

# Using inverse
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)

print("Solution using inverse:", X)

# Direct solve method (better method)
X2 = np.linalg.solve(A,B)
print("Solution using solve():", X2)
```

- Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

```python
size = 1000

# Python matrix multiplication
A = [[1]*size for _ in range(size)]
B = [[1]*size for _ in range(size)]

start = time.time()
result = [[sum(A[i][k]*B[k][j] for k in range(size))
           for j in range(size)] for i in range(size)]
print("Python matrix multiplication time:", time.time() - start)

# NumPy matrix multiplication
A_np = np.ones((size,size))
B_np = np.ones((size,size))

start = time.time()
result = np.dot(A_np, B_np)
print("NumPy matrix multiplication time:", time.time() - start)
```
